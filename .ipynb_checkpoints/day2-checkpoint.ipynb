```python
#day 2 part 1 rock paper scissors puzzle
file = open('day2input1.txt', 'r')
content = file.read()
games = str(content).strip().split('\n')
games
# AX AY AZ BX BY BZ CX CY CZ
```

```python
my_score = 0

for round in games:
    match round:
        #rock draw
        case "A X":
            my_score += 4
        #paper win    
        case "A Y":
            my_score += 8
        #scissors loss    
        case "A Z":
            my_score += 3
        #rock loss    
        case "B X":
            my_score += 1
        #paper draw    
        case "B Y":
            my_score += 5
        #scissors win    
        case "B Z":
            my_score += 9
        #rock win    
        case "C X":
            my_score += 7
        #paper loss    
        case "C Y":
            my_score += 2
        #scissors draw    
        case "C Z":
            my_score += 6
    
my_score
```
