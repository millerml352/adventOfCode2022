```python
def parse_stack_text(stacktext):
    stacks = [""]*10
    for line in stacktext[:-1]:
        for i, box in enumerate(line[1::4]):
            if box != " ": stacks[i+1] += box
    return stacks
```

```python
input_data = open("day5input1.txt").read()
stackt, instructions = [part.split("\n") for part in input_data.split("\n\n")]
stacks = parse_stack_text(stackt)

for command in instructions[:-1]:
    # print('**********')
    a, qty, b, origin, c, dest = command.split()
    qty = int(qty)
    origin = int(origin)
    dest = int(dest)
#     print('Stack: ' + stacks[origin])
    
#     print('Command: ' + command)
#     print('Quantity: ' + str(qty))
#     print('From: ' + str(origin))
#     print('To: ' + str(dest))
    
    i = 1
    while i <= qty:
        in_motion = stacks[origin][0]
        stacks[origin] = stacks[origin][1:]
        stacks[dest] = in_motion + stacks[dest]
        
        # print('-----')
        # print('Stack, after removal: ' + stacks[origin])
        # print('Crate taken: ' + in_motion)
        # print('Stack, after addition: ' + stacks[dest])
        
        i += 1
        
stacks
```

```python
final_arrangement = ''
for i,stack in enumerate(stacks):
    final_arrangement = final_arrangement + stack[0:1]
final_arrangement
```
